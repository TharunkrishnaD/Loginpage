```python
import tkinter as tk
from tkinter import messagebox
from PIL import ImageTk
import pymysql

    
def errormessage():
    if Name.get()=='' and Phone_Number.get()=='' and Password.get()=='':
        messagebox.showerror('Error','All field should be field')
    elif not Phone_Number.get().isdigit() or len(Phone_Number.get()) != 10:
        messagebox.showerror('Error', 'Phone number should be in Number')
    else:
        try:
            connection = pymysql.connect(host='localhost',user='root',password='9750528343')
            mycursor=connection.cursor()
        except:
            messagebox.showerror('Error','Database Connectivity Issue,Please try again')
            return
        try:
            query='create database userdata'
            mycursor.execute(query)
            query='use userdata'
            mycursor.execute(query)
            query='create table data(id int auto_increment primary key not null,user_name varchar(100) unique not null,phone_number varchar(20),gender varchar(10),password varchar(20))'
            mycursor.execute(query)
        except:
            mycursor.execute('use userdata')
        
        query='insert into data(user_name,phone_number,gender,password) values(%s,%s,%s,%s)'
        mycursor.execute(query,(Name.get(),Phone_Number.get(),gender.get(),Password.get()))
        connection.commit()
        connection.close()
        messagebox.showinfo('Success','Registration is successful')
        login_page()
def login_page():
    top.destroy()
    def welcome(uname,gen):
        login_window.destroy()
        top1 = tk.Tk()
        bgImage = ImageTk.PhotoImage(file="C:/Users/Hp/Desktop/Python task/tkinder/loginproject.jpg")
        top1.title("Log Page")
        top1.geometry("1000x1000")
        bgLabel = tk.Label(top1, image=bgImage)
        bgLabel.place(x=0, y=0)
        if gen == 'Male':
            label = tk.Label(top1, font=('calibri', 12, 'bold'), text=f"WELCOME Mr.{uname}")
        else:
            label = tk.Label(top1, font=('calibri', 12, 'bold'), text=f"WELCOME Miss/Mrs.{uname}")
        label.place(x=120, y=30)
        top1.mainloop()
    def login_user():
        if Name.get()=='' and Password.get()=='':
            messagebox.showerror('Error','All field should be field')
        else:
            try:
                connection = pymysql.connect(host='localhost',user='root',password='9750528343')
                mycursor=connection.cursor()
            except:
                messagebox.showerror('Error','Connection is not established try again')
                return
            query = 'use userdata'
            mycursor.execute(query)
            query = 'select * from data where user_name=%s and password=%s'
            mycursor.execute(query,(Name.get(),Password.get()))
            row=mycursor.fetchone()
            if row==None:
                messagebox.showerror('Error','Invalid username or password')
            else:
                messagebox.showinfo('Welcome','Login is sucessful')
                mycursor.execute(f"select gender from data where user_name='{Name.get()}'")
                result = mycursor.fetchone()
                if result:
                    gender = result[0]
                    welcome(Name.get(), gender)
                else:
                    messagebox.showerror('Error', 'Gender not found in the database')
    login_window = tk.Tk()
    bgImage = ImageTk.PhotoImage(file="C:/Users/Hp/Desktop/Python task/tkinder/loginproject.jpg")
    login_window.title("Login Page")
    login_window.geometry("1000x1000")
    bgLabel = tk.Label(login_window, image=bgImage)
    bgLabel.place(x=0, y=0)
    label = tk.Label(login_window, font=('calibri', 12, 'bold'), text="WELCOME")
    label.place(x=120, y=30)
    User_Name = tk.Label(login_window, text="Full Name")
    User_Name.place(x=40, y=60)
    User_Password = tk.Label(login_window, text="Password")
    User_Password.place(x=40, y=90)
    Name = tk.Entry(login_window, width=30)
    Name.place(x=110, y=60)
    Password = tk.Entry(login_window, width=30)
    Password.place(x=110, y=90)
    Sign_in = tk.Button(login_window, font=('calibri', 12, 'bold'), bg="yellow", text="Sign Up",command=login_user)
    Sign_in.place(x=150, y=130)
    login_window.mainloop()

top = tk.Tk()
bgImage = ImageTk.PhotoImage(file="C:/Users/Hp/Desktop/Python task/tkinder/loginproject.jpg")
top.title("Sign Up Page")
top.geometry("500x500")
gender = tk.StringVar(top, "Male")
bgLabel = tk.Label(top, image=bgImage)
bgLabel.place(x=0, y=0)
label = tk.Label(top, font=('calibri', 12, 'bold'), text="WELCOME")
label.place(x=120, y=30)

User_Name = tk.Label(top, text="Full Name")
User_Name.place(x=40, y=60)

User_Phone_Number = tk.Label(top, text="Mobile No")
User_Phone_Number.place(x=40, y=100)

User_Gender = tk.Label(top, text="Gender")
User_Gender.place(x=40, y=140)

User_Sign_in = tk.Label(top,text="Already have a account")
User_Sign_in.place(x=65, y=275)

User_radiobutton = tk.Radiobutton(top,text="Male", variable = gender,value="Male")
User_radiobutton.place(x=140, y=140)

User_radiobutton1 = tk.Radiobutton(top,text="Female",variable = gender,value="Female")
User_radiobutton1.place(x=220, y=140)

User_Password = tk.Label(top, text="Password")
User_Password.place(x=40, y=190)

Name = tk.Entry(top, width=30)
Name.place(x=110, y=60)

Phone_Number = tk.Entry(top, width=30)
Phone_Number.place(x=110, y=100)

Password = tk.Entry(top, width=30)
Password.place(x=110, y=190)

Sign_up = tk.Button(top, font=('calibri', 12, 'bold'), bg="yellow", text="Create Account",command= errormessage)
Sign_up.place(x=120, y=230)

Sign_in = tk.Button(top, font=('calibri', 12, 'bold'), bg="yellow", text="Sign Up",command=login_page)
Sign_in.place(x=220, y=270)

top.mainloop()
```
